```python
from src.config import settings
from pymilvus import MilvusClient
client = MilvusClient(
    uri = settings.ZILLIZ_ENDPOINT, 
    user = settings.ZILLIZ_USER, 
    password = settings.ZILLIZ_PASSWORD
)
```

```python
candidate_1 = {
    'chat_id': '34747507-0',
    'name': '王充',
}
candidate_2 = {
    'candidate_id': '461802494827847881',
    'name': '胡修齐',
}

print(client.query('CN_candidates',
    filter=f"chat_id == '{candidate_1['chat_id']}'"
))

print(client.query('CN_candidates',
    filter=f"name == '{candidate_1['name']}'"
))

candidate_2 = client.query('CN_candidates',
    filter=f"candidate_id == '{candidate_2['candidate_id']}'"
)[0]
print(candidate_2)
print(client.query('CN_candidates',
    filter=f"name LIKE '{candidate_2['name']}%'"
))
```

```python
c = candidate_2.copy()
c.pop('resume_vector')
c.pop('resume_text')
c.pop('analysis')
c.pop('last_message')
c.pop('metadata')
c.pop('chat_id')
c.pop('updated_at')
c.pop('conversation_id')
# c.pop('name')
c = {k: v for k, v in c.items() if v}
print(c)
result = client.upsert('CN_candidates', data=c, partial_update=True)
result['ids']
```
