# Daily Reports Test Notebook

This notebook tests the daily report functionality:
- JSON statistics API response
- Formatted report (Markdown) response
- Report formatting functions


```python
import json
import requests

# Configuration
TEST_MODE = "local"  # Options: "local" or "remote"
LOCAL_URL = "http://localhost:3000"
REMOTE_URL = ""  # Set your Vercel deployment URL here

BASE_URL = LOCAL_URL if TEST_MODE == "local" else REMOTE_URL
print(f"Testing against: {BASE_URL}")
```

## Test 1: Import and Test Stats Functions


```python
response = requests.get(f"{BASE_URL}/api/stats", params={"format": "report"})
overall_report = response.json()
print("keys:", overall_report.keys())
print("Title:", overall_report.get("title"))
print("\nMessage:")
print(overall_report.get("message"))
```

## Test 2: Get Statistics Data


```python
response = requests.get(f"{BASE_URL}/api/send-report", params={"type": "overall"})
result = response.json()

print(json.dumps(result, indent=2, ensure_ascii=False))
```

```python
# Get job index (0-based)
job_index = 1  # Change this to test different jobs

response = requests.get(f"{BASE_URL}/api/stats", params={"format": "job_report", "job_index": job_index})
job_report = response.json()

print("Title:", job_report.get("title"))
print("\nMessage:")
print(job_report.get("message"))
```

## Test 3: Send report

```python
# Use the same job_index as above
response = requests.get(f"{BASE_URL}/api/send-report", params={"type": "job", "job_index": job_index})
result = response.json()

print(json.dumps(result, indent=2, ensure_ascii=False))

# Show warning if using fallback
if result.get("warning"):
    print(f"\nâš ï¸ {result.get('warning')}")
```
